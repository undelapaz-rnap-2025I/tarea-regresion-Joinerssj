```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Cargar datos (1 característica + 1 salida)
#    Se asume un archivo CSV con dos columnas: 
#    - columna 0: la única característica (p. ej. tamaño de población en decenas de miles)
#    - columna 1: el valor objetivo (p. ej. beneficio en 10.000$)
data = np.loadtxt('ex1data1.txt', delimiter=',')
X_original = data[:, 0].reshape(-1, 1)  # matriz (m×1) con la característica
y          = data[:, 1].reshape(-1, 1)  # vector columna (m×1) con los valores de salida

# 2) Normalización de características
def featureNormalize(X):
    """
    Normaliza cada columna de X de forma que tenga media 0 y desviación estándar 1.
    Devuelve:
      - X_norm: matriz normalizada
      - mu:     vector con las medias originales
      - sigma:  vector con las desviaciones estándar originales
    """
    mu    = np.mean(X, axis=0)   # media de cada columna (1,)
    sigma = np.std(X, axis=0)    # desviación estándar de cada columna (1,)
    X_norm = (X - mu) / sigma    # aplicar normalización
    return X_norm, mu, sigma

# Aplicar normalización a los datos originales
X_norm, mu, sigma = featureNormalize(X_original)

# 3) Añadir término de sesgo (bias)
#    Creamos una columna de 1s para el parámetro θ₀
m, n = X_norm.shape                                  # m = nº ejemplos, n = nº features (1)
X_with_bias = np.hstack([np.ones((m,1)), X_norm])    # matriz (m × 2)

# 4) Función de coste para regresión lineal
def cost_function(theta, X, y):
    """
    Calcula J(θ) = (1/(2m)) * sum((X·θ - y)^2)
    donde:
      - theta: vector de parámetros (2×1)
      - X:     matriz de características con bias (m×2)
      - y:     vector de salidas (m×1)
    """
    m     = len(y)
    preds = X @ theta            # hipótesis (m×1)
    sq_e  = (preds - y)**2       # errores al cuadrado
    return (1/(2*m)) * np.sum(sq_e)

# 5) Descenso de gradiente
def gradientDescent(X, y, theta, alpha, iterations):
    """
    Realiza descensos de gradiente para ajustar theta.
    Devuelve:
      - theta: parámetros optimizados
      - J_history: array con el coste en cada iteración
    """
    m         = len(y)
    J_history = np.zeros(iterations)
    for it in range(iterations):
        preds  = X @ theta                  # (m×1) hipótesis actual
        errors = preds - y                  # (m×1) errores
        grad_J = (1/m) * (X.T @ errors)     # (2×1) gradiente de J respecto a θ
        theta  = theta - alpha * grad_J     # actualizar parámetros
        J_history[it] = cost_function(theta, X, y)  
        # Mostrar iteración, coste y gradiente (convertido a lista para legibilidad)
        print(f"Iter {it+1:3d} | Cost: {J_history[it]:.6f} | Grad: {grad_J.ravel().tolist()}")
    return theta, J_history

# 6) Inicializar parámetros y ejecutar descenso de gradiente
theta_init = np.zeros((n+1, 1))   # vector (2×1) inicial de ceros
alpha      = 0.01                 # tasa de aprendizaje
iterations = 300                  # nº de iteraciones

print("Antes de GD:", X_with_bias.shape, theta_init.shape)
theta, J_hist = gradientDescent(X_with_bias, y, theta_init, alpha, iterations)
print("Theta final:", theta.ravel())

# 7) Graficar evolución del coste
plt.figure()
plt.plot(np.arange(1, iterations+1), J_hist)
plt.xlabel('Iteraciones')
plt.ylabel('Coste J')
plt.title('Descenso de gradiente: coste vs. iteraciones')
plt.show()

# 8)

```
