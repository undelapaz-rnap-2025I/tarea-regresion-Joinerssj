```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Cargar datos (1 caracterstica + 1 salida)
#    Se asume un archivo CSV con dos columnas: 
#    - columna 0: la nica caracterstica (p. ej. tamao de poblacin en decenas de miles)
#    - columna 1: el valor objetivo (p. ej. beneficio en 10.000$)
data = np.loadtxt('ex1data1.txt', delimiter=',')
X_original = data[:, 0].reshape(-1, 1)  # matriz (m1) con la caracterstica
y          = data[:, 1].reshape(-1, 1)  # vector columna (m1) con los valores de salida

# 2) Normalizacin de caractersticas
def featureNormalize(X):
    """
    Normaliza cada columna de X de forma que tenga media 0 y desviacin estndar 1.
    Devuelve:
      - X_norm: matriz normalizada
      - mu:     vector con las medias originales
      - sigma:  vector con las desviaciones estndar originales
    """
    mu    = np.mean(X, axis=0)   # media de cada columna (1,)
    sigma = np.std(X, axis=0)    # desviacin estndar de cada columna (1,)
    X_norm = (X - mu) / sigma    # aplicar normalizacin
    return X_norm, mu, sigma

# Aplicar normalizacin a los datos originales
X_norm, mu, sigma = featureNormalize(X_original)

# 3) Aadir trmino de sesgo (bias)
#    Creamos una columna de 1s para el parmetro 
m, n = X_norm.shape                                  # m = n ejemplos, n = n features (1)
X_with_bias = np.hstack([np.ones((m,1)), X_norm])    # matriz (m  2)

# 4) Funcin de coste para regresin lineal
def cost_function(theta, X, y):
    """
    Calcula J() = (1/(2m)) * sum((X - y)^2)
    donde:
      - theta: vector de parmetros (21)
      - X:     matriz de caractersticas con bias (m2)
      - y:     vector de salidas (m1)
    """
    m     = len(y)
    preds = X @ theta            # hiptesis (m1)
    sq_e  = (preds - y)**2       # errores al cuadrado
    return (1/(2*m)) * np.sum(sq_e)

# 5) Descenso de gradiente
def gradientDescent(X, y, theta, alpha, iterations):
    """
    Realiza descensos de gradiente para ajustar theta.
    Devuelve:
      - theta: parmetros optimizados
      - J_history: array con el coste en cada iteracin
    """
    m         = len(y)
    J_history = np.zeros(iterations)
    for it in range(iterations):
        preds  = X @ theta                  # (m1) hiptesis actual
        errors = preds - y                  # (m1) errores
        grad_J = (1/m) * (X.T @ errors)     # (21) gradiente de J respecto a 
        theta  = theta - alpha * grad_J     # actualizar parmetros
        J_history[it] = cost_function(theta, X, y)  
        # Mostrar iteracin, coste y gradiente (convertido a lista para legibilidad)
        print(f"Iter {it+1:3d} | Cost: {J_history[it]:.6f} | Grad: {grad_J.ravel().tolist()}")
    return theta, J_history

# 6) Inicializar parmetros y ejecutar descenso de gradiente
theta_init = np.zeros((n+1, 1))   # vector (21) inicial de ceros
alpha      = 0.01                 # tasa de aprendizaje
iterations = 300                  # n de iteraciones

print("Antes de GD:", X_with_bias.shape, theta_init.shape)
theta, J_hist = gradientDescent(X_with_bias, y, theta_init, alpha, iterations)
print("Theta final:", theta.ravel())

# 7) Graficar evolucin del coste
plt.figure()
plt.plot(np.arange(1, iterations+1), J_hist)
plt.xlabel('Iteraciones')
plt.ylabel('Coste J')
plt.title('Descenso de gradiente: coste vs. iteraciones')
plt.show()

# 8)

```
